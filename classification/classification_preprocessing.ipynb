# Classification preprocessing


Read json data directly with pandas

```python
import pandas as pd
import numpy as np
json_file_path = 'Data-Test2.json'
df = pd.read_json(json_file_path, convert_dates=True)
```

Transpose the dataframe:

```python
dfT = df.T
dfT.head()
```

The dictionary situated on the first line and first column.

```python
dfT.iloc[0,[0]].values
```

Creation of a complete dataframe.

```python
climb = dfT[['climb']]
first_value_for_climb = climb.iloc[0,[0]].values[0]
dfmerge = pd.DataFrame(first_value_for_climb, index=[climb.iloc[0,[0]].name])
for i in range(1,dfT.shape[0]):
    value_for_climb = climb.iloc[i,[0]].values[0]
    new_df_cli = pd.DataFrame(value_for_climb, index=[climb.iloc[i,[0]].name])
    dfmerge = pd.concat([dfmerge,new_df_cli])
dfmerge_renamed_climb=dfmerge.rename(columns = {dfmerge.columns[i]:dfmerge.columns[i] + '_climb' for i in range(dfmerge.shape[1])})

cruise = dfT[['cruise']]
first_value_for_cruise = cruise.iloc[0,[0]].values[0]
dfmerge_cruise = pd.DataFrame(first_value_for_cruise, index=[cruise.iloc[0,[0]].name])
for i in range(1,dfT.shape[0]):
    value_for_cruise = cruise.iloc[i,[0]].values[0]
    new_df_cru = pd.DataFrame(value_for_cruise, index=[cruise.iloc[i,[0]].name])
    dfmerge_cruise = pd.concat([dfmerge_cruise,new_df_cru])
dfmerge_renamed_cruise=dfmerge_cruise.rename(columns = {dfmerge_cruise.columns[i]:dfmerge_cruise.columns[i] + '_cruise' for i in range(dfmerge_cruise.shape[1])})

descent = dfT[['descent']]
first_value_for_descent = descent.iloc[0,[0]].values[0]
dfmerge_descent = pd.DataFrame(first_value_for_descent, index=[descent.iloc[0,[0]].name])
for i in range(1,dfT.shape[0]):
    value_for_descent = descent.iloc[i,[0]].values[0]
    new_df_des = pd.DataFrame(value_for_descent, index=[descent.iloc[i,[0]].name])
    dfmerge_descent = pd.concat([dfmerge_descent,new_df_des])
dfmerge_renamed_descent=dfmerge_descent.rename(columns = {dfmerge_descent.columns[i]:dfmerge_descent.columns[i]+ '_descent' for i in range(dfmerge_descent.shape[1])})

df_final = pd.concat([dfmerge_renamed_climb,dfmerge_renamed_cruise,dfmerge_renamed_descent],axis=1)
df_final
```

For this dataframe, the index is ICAO24_callsign, according to which one can figure out which airline the aircraft is belong to.
